## One Hot Encoding


```python
import pandas as pd
```

```python
df=pd.DataFrame({"Class":['M','F','M','M','F','F','M','F','F','M'],
                  "age":[23,22,26,23,27,29,31,24,33,35]})
df
```

```python
y = pd.get_dummies(df.Class, prefix='Gender')
print(y.head())
```

```python
data=pd.read_csv(r"C:\Users\DELL\Desktop\iris.csv")
data
```

```python
y = pd.get_dummies(data.Species, prefix='Species')
print(y.tail())
```

```python
data=data['Species']
data=pd.DataFrame(data)
data
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
y = LabelBinarizer().fit_transform(data.Species)
y
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
y = OneHotEncoder().fit_transform(data).toarray()
print(y)
```
