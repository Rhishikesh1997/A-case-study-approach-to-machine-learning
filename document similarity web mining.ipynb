```python
import gensim
```

```python
import glob
file_names=glob.glob("stories/*.txt")
file_names
```

```python
raw_documents=[]
for file in file_names:
    try:#we will try to catch block to prevent the code from crashing if it can not rewd any of the text
        with open(file,"r",encoding="utf-8") as f:raw_documents.append(f.read())
    except:
        pass

```

```python
print("Number of documents:",len(raw_documents))
```

```python
clean_texts=[]
for text in raw_documents:
    clean_texts.append(gensim.utils.simple_preprocess(text))
```

```python
dictionary=gensim.corpora.Dictionary(clean_texts)
print("Number of words in dictionary:",len(dictionary))
for i in range(100) :
    print(i,dictionary[i])
```

```python
corpus=[dictionary.doc2bow(text) for text in clean_texts]
print(corpus[:10])
```

```python
tf_idf=gensim.models.TfidfModel(corpus)
print(tf_idf)
```

```python
similarity_object=gensim.similarities.Similarity('stories',tf_idf[corpus],num_features=len(dictionary))
print(similarity_object)
print(type(similarity_object))
```

```python
text=raw_documents[8]
query_doc=gensim.utils.simple_preprocess(text)
print(query_doc)
```

```python
query_doc_bow=dictionary.doc2bow(query_doc)
print(query_doc_bow)
```

```python
query_doc_tf_idf=tf_idf[query_doc_bow]
print(query_doc_tf_idf)
```

```python
similarity_scores=list(similarity_object[query_doc_tf_idf])
print(similarity_scores)
```

```python
max_score=max(similarity_scores)
print(max_score)
similarity_scores.index(max_score)
```
