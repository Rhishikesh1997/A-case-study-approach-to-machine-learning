```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.manifold import TSNE
```

```python
data=pd.read_csv(r"C:\Users\DELL\Desktop\iris.csv")
data.head(2)
```

```python
train=data.iloc[:,1:5].values
train
train.reshape(1,-1)
```

```python
target=data.iloc[:,5]
target
```

```python
from sklearn.preprocessing import LabelEncoder
number=LabelEncoder()
target=number.fit_transform(target.astype('str'))
target
```

```python
scaler=StandardScaler()
data_x=scaler.fit_transform(train)
data_x
```

```python
model=TSNE(n_components=2,random_state=0,perplexity=35.0,angle=0,n_iter=2000)
tsne_data=model.fit_transform(data_x)
print(tsne_data)
```

```python
tsne_df=pd.DataFrame(data=tsne_data)
tsne_df
```

# 2D visualization of t-SNE plot

```python
import matplotlib.pyplot as plt
plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=target)
plt.title("t-SNE plot of Iris Dataset",fontsize=20)
```

# 3D visualization of t-SNE plot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(19,15))
ax=fig.add_subplot(111,projection='3d')
ax.scatter (tsne_data[:,1],tsne_data[:,0],c=target,s=10)
ax.view_init(30,185)
```
